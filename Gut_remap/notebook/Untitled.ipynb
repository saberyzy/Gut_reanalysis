```python
def cell_frac_2(adata):
    df = adata.obs.groupby(['leiden_r1','patient_id'], observed=False).size().unstack()
    df = df.transpose()
    df['cell_total_number'] = df.sum(axis = 1)
    n_leiden_r1 = len(adata.obs['leiden_r1'].unique())
    for c in df.iloc[:,0:n_leiden_r1].columns:
        name = c + ' frac'
        df[name] = df[c] / df['cell_total_number']

    for group in adata.obs['Group'].unique():
        for ID in adata.obs.loc[adata.obs['Group'] == group, 'patient_id'].unique():
            df.loc[ID, 'group'] = group
            
    
    df.group = pd.Categorical(df.group, 
                      categories=['HC', 'IT', 'IA', 'AR', 'CR'],
                      ordered=True)
    
    df = df.sort_values(by = ['group'])
    
    
    
    return(df)
```

```python
adata_CD8_l = adata_CD8[adata_CD8.obs.tissue == 'Liver'].copy()
```

```python
df = cell_frac_2(adata_CD8_l)
```

```python
def scatterplot(df, group, celltype, palette,
                xlim, ylim, title, ylabel, path, name):
    rcParams['pdf.fonttype'] = 42
    rcParams['font.size'] = 16
    rcParams['figure.figsize'] = (4,7)
    
    
    fig, ax1 = plt.subplots(1,1)
    
    sb.scatterplot(x=group, y=celltype, hue = group, palette = palette, data=df, 
                   s = 300, alpha = 0.9, marker="$\circ$", ec="face", ax=ax1)
    
    PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'black'},
    'medianprops':{'color':'black'},
    'whiskerprops':{'color':'black'}
    }
    sb.boxplot(x=group,y=celltype,
            data=df,
            showfliers=False,
            linewidth=1.5, 
            **PROPS, ax = ax1)
    
    
    
#     ax1.set_xticklabels(['Naïve''\n''(n=5)', 'AAV''\n''(n=5)', 'AAV-HBV''\n''(n=5)'])
    ax1.set_xlim(xlim)
    ax1.set_ylim(ylim)
    ax1.set_title(title)
    ax1.set_xlabel("")
    ax1.set_ylabel(ylabel)

    ax1.legend(title = 'Group', bbox_to_anchor=(1.5, 1))
    
    #IT vs HC
#     ax1.arrow(-0.2,0.3, 1.5,0,head_length=0.0,head_width=0.0, width = 0.001)
#     ax1.text(-0.3, 0.32, 'p = 0.02', fontsize = 16)
    #IA vs HC
#     ax1.arrow(-0.2,0.4, 2.5,0,head_length=0.0,head_width=0.0, width = 0.001)
#     ax1.text(0.3, 0.43, 'p = 0.063', fontsize = 16)
#     #IA vs IT
#     ax1.arrow(0.8,0.6, 1.5,0,head_length=0.0,head_width=0.0, width = 0.001)
#     ax1.text(0.9, 0.63, 'p = 0.041', fontsize = 16)
#     #AR vs HC
#     ax1.arrow(-0.2,1.1, 3.5,0,head_length=0.0,head_width=0.0, width = 0.001)
#     ax1.text(0.9, 1.12, 'p = 0.012', fontsize = 16)
#     #CR vs HC
#     ax1.arrow(-0.2,1.3, 4.5,0,head_length=0.0,head_width=0.0, width = 0.001)
#     ax1.text(1.1, 1.32, 'p = 0.024', fontsize = 16)
    
    save_dir = os.path.join(path, name)
    fig.savefig(save_dir, bbox_inches='tight')
```
