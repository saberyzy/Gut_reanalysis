```python
#generate the data for only one disease group, test for different diease group
def cpdb_prep(adata_t, group, root_path):
    temp = adata_t[adata_t.obs.Group == group].copy()
    sc.pp.filter_genes(temp, min_cells=3)
    sc.pp.normalize_total(temp, target_sum=1e4)

    df = temp.obs.loc[:,['subtype']].copy()
    df.index.name = 'Cell'
    df.columns = ['cell_type']
    
    temp_c = temp.X.toarray()
    count = pd.DataFrame(data=temp_c, index=temp.obs.index, columns=list(temp.var.gene_ids))
    
    count = count.transpose()
    count.index.name = 'Gene'
    count.columns.name = None
    
    savepath = os.path.join(root_path, group)
    os.mkdir(savepath)
    
    m_path = os.path.join(savepath, 'meta_T.tsv')
    df.to_csv(m_path, sep = '\t')
    c_path = os.path.join(savepath, 'counts_T.h5')
    count.to_hdf(c_path, key='df')
    
def cpdb_batch(adata_t, group_list, root_path):
    for group in group_list:
        print('generate group {} files...'.format(group))
        cpdb_prep(adata_t, group, root_path)
        print('{} group files generation completed'.format(group))
        
    
    
    
    
    
    
```

```python
root_path = '/domino/edv/id-td-virology/Zhiyuan/public/Gut_remap/processed_data/cellphonedb/'
group_list = ['IA', 'HC', 'CR', 'AR']

cpdb_batch(adata_t, group_list, root_path)
```
