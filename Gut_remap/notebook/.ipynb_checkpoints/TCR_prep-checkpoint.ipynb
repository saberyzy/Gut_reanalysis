```python
temp = df_2.groupby(['patientID', 
                        'TRBV']).size().to_frame(name = 'counts').reset_index().copy()
df_plot = temp.sort_values(by = ['patientID', 'counts'],ascending = False).reset_index(drop = True)


df_plot['frac'] = 0
df_plot['TRBV_plot'] = df_plot['TRBV'].copy()
others_counts_sum = 0
patient_list = df_plot.patientID.unique()

for ID in patient_list:
    ID_sum = df_plot.loc[df_plot.patientID == ID, 'counts'].sum()
    ind_list = df_plot.loc[df_plot.patientID == ID].index
    for ind in ind_list:
        df_plot.loc[ind,'frac'] = df_plot.loc[ind, 'counts'] / ID_sum
        if df_plot.loc[ind, 'frac'] < 0.01:
            if others_counts_sum == 0:
                df_plot.loc[ind, 'TRBV_plot'] = 'others'
#                 print(df_plot.loc[ind,:])
                others_counts_sum = others_counts_sum + df_plot.loc[ind, 'counts']
            else:
                others_counts_sum = others_counts_sum + df_plot.loc[ind, 'counts']
                df_plot.drop(index=[ind], inplace = True)
            
    df_plot.loc[(df_plot.patientID == ID)&(df_plot.TRBV_plot == 'others'), 'counts'] = others_counts_sum
    df_plot.loc[(df_plot.patientID == ID)&(df_plot.TRBV_plot == 'others'), 'frac'] = others_counts_sum/ID_sum
    #refresh parameters, next patient
    others_counts_sum = 0

        








        
    
    
```

```python
df_plot_2 = df_plot.pivot(index='patientID', columns='TRBV_plot', values='frac').fillna(0)
```

```python
def stackplot(df, color_list, path, name):
    rcParams['pdf.fonttype'] = 42
    rcParams['font.size'] = 16
    rcParams['figure.figsize'] = (14,4)
    
    fig, ax1 = plt.subplots(1,1)
    color = color_list[0:len(df.columns)]
    df.plot(kind='bar',stacked = True, width=0.9, color = color,ax = ax1)

    ax1.set_xlabel("")
    ax1.set_ylabel("TRBV Fraction")  
    ax1.set_title("TRBV Fraction in each patient") 
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=60,ha = 'right')
    ax1.legend(title = 'TRBV gene names', ncol=3,bbox_to_anchor=(1, 1))
    
    save_dir = os.path.join(path, name)
    fig.savefig(save_dir, bbox_inches='tight')
```

```python
def stackplot_3(df, color_list, path, name):
    rcParams['pdf.fonttype'] = 42
    rcParams['font.size'] = 16
    rcParams['figure.figsize'] = (14,4)
    
    fig, ax1 = plt.subplots(1,1)
    color = color_list[0:len(df.columns)]
    df.plot(kind='bar',stacked = True, width=0.9, color = color,ax = ax1)

    ax1.set_xlabel("")
    ax1.set_ylabel("patient counts")  
    ax1.set_title("Patient numbers detected for each clonal TRBV") 
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90,ha = 'center')
    ax1.legend(title = 'Disease group', ncol=1,bbox_to_anchor=(1, 1))
    
    save_dir = os.path.join(path, name)
    fig.savefig(save_dir, bbox_inches='tight')
```
