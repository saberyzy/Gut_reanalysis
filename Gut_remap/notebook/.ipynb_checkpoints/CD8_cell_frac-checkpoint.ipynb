```python
#CD8 CELL FRAC
import scanpy as sc #for scanpy >= 1.3.7
import anndata as ann
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib import colors
import seaborn as sb
import logging
import os
import glob
import matplotlib
import math
```

```python
savepath = '/domino/edv/id-td-virology/Zhiyuan/public/Gut_remap/processed_data/CD8.h5ad'
adata_CD8 = sc.read_h5ad(savepath)
```

```python
def cell_frac(adata):
    df = adata.obs.groupby(['subtype','patient_id'], observed=False).size().unstack()
    df = df.transpose()
    df['cell_total_number'] = df.sum(axis = 1)
    n_subtype = len(adata.obs['subtype'].unique())
    for c in df.iloc[:,0:n_subtype].columns:
        name = c + ' frac'
        df[name] = df[c] / df['cell_total_number']

    for group in adata.obs['Group'].unique():
        for ID in adata.obs.loc[adata.obs['Group'] == group, 'patient_id'].unique():
            df.loc[ID, 'group'] = group
            
    
    df.group = pd.Categorical(df.group, 
                      categories=['HC', 'IT', 'IA', 'AR', 'CR'],
                      ordered=True)
    
    df = df.sort_values(by = ['group'])
    
    
    
    return(df)

```

```python
def scatterplot(df, group, celltype, palette,
                xlim, ylim, title, ylabel, path, name):
    rcParams['pdf.fonttype'] = 42
    rcParams['font.size'] = 16
    rcParams['figure.figsize'] = (4,7)
    
    
    fig, ax1 = plt.subplots(1,1)
    
    sb.scatterplot(x=group, y=celltype, hue = group, palette = palette, data=df, 
                   s = 300, alpha = 0.9, marker="$\circ$", ec="face", ax=ax1)
    
    PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'black'},
    'medianprops':{'color':'black'},
    'whiskerprops':{'color':'black'}
    }
    sb.boxplot(x=group,y=celltype,
            data=df,
            showfliers=False,
            linewidth=1.5, 
            **PROPS, ax = ax1)
    
    
    
#     ax1.set_xticklabels(['Naïve''\n''(n=5)', 'AAV''\n''(n=5)', 'AAV-HBV''\n''(n=5)'])
    ax1.set_xlim(xlim)
    ax1.set_ylim(ylim)
    ax1.set_title(title)
    ax1.set_xlabel("")
    ax1.set_ylabel(ylabel)

    ax1.legend(title = 'Group', bbox_to_anchor=(1, 1))
    
    #IT vs HC
    ax1.arrow(-0.2,0.8, 1.5,0,head_length=0.0,head_width=0.0, width = 0.001)
    ax1.text(-0.3, 0.82, 'p = 0.0076', fontsize = 16)
    #IA vs HC
    ax1.arrow(-0.2,0.9, 2.5,0,head_length=0.0,head_width=0.0, width = 0.001)
    ax1.text(0.3, 0.93, 'p = 0.0043', fontsize = 16)
    #IA vs IT
    ax1.arrow(0.8,0.6, 1.5,0,head_length=0.0,head_width=0.0, width = 0.001)
    ax1.text(0.9, 0.63, 'p = 0.041', fontsize = 16)
    #AR vs HC
    ax1.arrow(-0.2,1.1, 3.5,0,head_length=0.0,head_width=0.0, width = 0.001)
    ax1.text(0.9, 1.12, 'p = 0.012', fontsize = 16)
    #CR vs HC
    ax1.arrow(-0.2,1.3, 4.5,0,head_length=0.0,head_width=0.0, width = 0.001)
    ax1.text(1.1, 1.32, 'p = 0.024', fontsize = 16)
    
    save_dir = os.path.join(path, name)
    fig.savefig(save_dir, bbox_inches='tight')
```

```python
def dotplot(adata, path, marker_genes, title, group, c_order, key, name):
#     rcParams['figure.figsize']=(14,7)
    rcParams['pdf.fonttype'] = 42
    rcParams['font.size'] = 18
    
    temp = adata.copy()

#     sc.pl.dotplot(temp, marker_genes, groupby='cell_subtype_new', use_raw=False, figsize = (14, 7),
#                   save = 'dotplot.pdf')
    dp = sc.pl.rank_genes_groups_dotplot(temp, 
                                         title = title,
                                         var_names=marker_genes, 
                                         key=key,
                                         groupby=group,
                                         values_to_plot=None,
                                         dendrogram = False,
                                         cmap='Reds',
                                         use_raw=False, 
#                                          figsize = (8, 3),
                                         categories_order = c_order,
                                         swap_axes = False,
                                         return_fig=True,
                                         show = False)
    dp.legend(width = 2.5, colorbar_title='Mean expression').savefig(name)
```

```python
#plot Tem results in liver
def scatterplot_6(df, group, celltype, palette,
                xlim, ylim, title, ylabel, path, name):
    rcParams['pdf.fonttype'] = 42
    rcParams['font.size'] = 16
    rcParams['figure.figsize'] = (4,7)
    
    
    fig, ax1 = plt.subplots(1,1)
    
    sb.scatterplot(x=group, y=celltype, hue = group, palette = palette, data=df, 
                   s = 300, alpha = 0.9, marker="$\circ$", ec="face", ax=ax1)
    
    PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'black'},
    'medianprops':{'color':'black'},
    'whiskerprops':{'color':'black'}
    }
    sb.boxplot(x=group,y=celltype,
            data=df,
            showfliers=False,
            linewidth=1.5, 
            **PROPS, ax = ax1)
    
    
    
#     ax1.set_xticklabels(['Naïve''\n''(n=5)', 'AAV''\n''(n=5)', 'AAV-HBV''\n''(n=5)'])
    ax1.set_xlim(xlim)
    ax1.set_ylim(ylim)
    ax1.set_title(title)
    ax1.set_xlabel("")
    ax1.set_ylabel(ylabel)

    ax1.legend(title = 'Group', bbox_to_anchor=(1, 1))
    
    #IT vs HC
    ax1.arrow(-0.2,0.7, 1.5,0,head_length=0.0,head_width=0.0, width = 0.001)
    ax1.text(-0.3, 0.72, 'p = 0.021', fontsize = 16)
    #IA vs HC
    ax1.arrow(-0.2,0.85, 2.5,0,head_length=0.0,head_width=0.0, width = 0.001)
    ax1.text(0.3, 0.87, 'p = 0.088', fontsize = 16)
#     #IA vs IT
#     ax1.arrow(0.8,0.6, 1.5,0,head_length=0.0,head_width=0.0, width = 0.001)
#     ax1.text(0.9, 0.63, 'p = 0.041', fontsize = 16)
#     AR vs HC
    ax1.arrow(-0.2,1, 3.5,0,head_length=0.0,head_width=0.0, width = 0.001)
    ax1.text(0.9, 1.02, 'p = 0.048', fontsize = 16)
    #CR vs HC
#     ax1.arrow(-0.2,1.3, 4.5,0,head_length=0.0,head_width=0.0, width = 0.001)
#     ax1.text(1.1, 1.32, 'p = 0.024', fontsize = 16)

#     #AR vs CR
#     ax1.arrow(2.6,0.8, 1.5,0,head_length=0.0,head_width=0.0, width = 0.001)
#     ax1.text(2.8, 0.82, 'p = 0.05', fontsize = 16)
    
    save_dir = os.path.join(path, name)
    fig.savefig(save_dir, bbox_inches='tight')
```
