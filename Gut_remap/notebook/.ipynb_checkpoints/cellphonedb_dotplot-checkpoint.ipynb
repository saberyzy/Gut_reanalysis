```python
#build a custom dotplot for interpret cellphonedb result
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib import colors
import seaborn as sb
import pandas as pd
import numpy as np
import os
from ast import literal_eval
```

```python
def table_gen(root_path, group_list, pair_list, cell_pair):
    dic = {}
    for item in ['out/significant_means.txt', 'out/pvalues.txt']:
        df = pd.DataFrame(0, index = pair_list, columns = group_list)
        for group in group_list:
            path = os.path.join(root_path,group, item) 
            temp = pd.read_table(path)
            temp.set_index('interacting_pair', inplace = True)
            temp = temp.loc[pair_list, cell_pair].to_frame().copy()
            if item == 'out/significant_means.txt':
                temp = temp.fillna(0)
                temp.loc[:,'log_mean'] = np.log1p(temp.loc[:,cell_pair])
                df.loc[pair_list,group] = temp.loc[pair_list, 'log_mean']
            else:
                temp = temp.fillna(1)
                df.loc[pair_list,group] = temp.loc[pair_list, cell_pair]
        if item == 'out/significant_means.txt':
            dic['log_mean'] = df
        else:
            dic['p_value'] = df
    dic['cell_pair'] = cell_pair
    return(dic)

        
        
```

```python
def check_colornorm(vmin=None, vmax=None, vcenter=None, norm=None):
    from matplotlib.colors import Normalize

    try:
        from matplotlib.colors import TwoSlopeNorm as DivNorm
    except ImportError:
        # matplotlib<3.2
        from matplotlib.colors import DivergingNorm as DivNorm

    if norm is not None:
        if (vmin is not None) or (vmax is not None) or (vcenter is not None):
            raise ValueError('Passing both norm and vmin/vmax/vcenter is not allowed.')
    else:
        if vcenter is not None:
            norm = DivNorm(vmin=vmin, vmax=vmax, vcenter=vcenter)
        else:
            norm = Normalize(vmin=vmin, vmax=vmax)

    return norm

def dotplot(ax, df_size, df_color, cell_pair):
    mean_flat = df_color.values.flatten()
    y, x = np.indices(df_color.shape)
    y = y.flatten() + 0.5
    x = x.flatten() + 0.5
    normalize = check_colornorm()
    cmap = plt.get_cmap('viridis')
    color = cmap(normalize(mean_flat))
    
    frac = df_size.values.flatten()
    dot_max = np.ceil(max(frac) * 10) / 10
    dot_min = 0
    old_range = dot_max - dot_min
    frac = (frac - dot_min) / old_range
    size_exponent = 2
    size = frac**size_exponent
    largest_dot = 200
    smallest_dot = 0
    size = size * (largest_dot - smallest_dot) + smallest_dot
    
#     rcParams['figure.figsize'] = (4,10)
#     fig,ax = plt.subplots()
    
    ax.scatter(x,
               y,
               s=size,
               cmap=cmap,
               c=color,
               norm=normalize)
    
    
    y_ticks = np.arange(df_color.shape[0]) + 0.5
    ax.set_yticks(y_ticks)
    ax.set_yticklabels(
    [df_color.index[idx] for idx, _ in enumerate(y_ticks)], minor=False
    )

    x_ticks = np.arange(df_color.shape[1]) + 0.5
    ax.set_xticks(x_ticks)
    ax.set_xticklabels(
            [df_color.columns[idx] for idx, _ in enumerate(x_ticks)],
            rotation=0,
            ha='center',
            minor=False,
    )
    ax.tick_params(axis='both', labelsize=20)
    ax.grid(False)
    ax.set_ylabel('')
    
    x_padding = 0
    y_padding = 0
    x_padding = x_padding + 0.5
    y_padding = y_padding + 0
    ax.set_ylim(-y_padding, df_color.shape[0] + y_padding)
    ax.set_xlim(-x_padding, df_color.shape[1] + x_padding)
    
    return(dot_max, dot_min, largest_dot, smallest_dot, normalize)
    
def plot_size_legend(ax, dot_min, dot_max, largest_dot, smallest_dot):
    
    diff = dot_max - dot_min
    if diff < 1:
        step = 0.2
    else:
        step = 1
        
    size_range = np.arange(dot_max, dot_min, step * -1)[::-1]
    dot_range = dot_max - dot_min
    size_values = (size_range - dot_min) / dot_range
    
    size_exponent = 2
    size = size_values**size_exponent
    size = size * (largest_dot - smallest_dot) + smallest_dot

    # plot size bar
    ax.scatter(
        np.arange(len(size)) + 0.5,
        np.repeat(0, len(size)),
        s=size,
        color='gray',
        edgecolor='black',
        zorder=100,
    )
    ax.set_xticks(np.arange(len(size)) + 0.5)
    labels = [
        "{}".format(np.round((x), decimals=2)) for x in size_range
    ]
    ax.set_xticklabels(labels, fontsize=24)

    # remove surrounding lines
    ax.tick_params(axis='y', left=False, labelleft=False, labelright=False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.grid(False)

    ax.set_ylim(- largest_dot * 0.003, largest_dot * 0.003)
    ymax = ax.get_ylim()[1]
    ax.set_title('-log10(p-value)', y = ymax+0.3, size=20)

    xmin, xmax = ax.get_xlim()
    ax.set_xlim(xmin - 0.15, xmax + 0.5)    

def plot_colorbar(color_legend_ax, normalize):
    """
    Plots a horizontal colorbar given the ax an normalize values
    Parameters
    ----------
    color_legend_ax
    normalize
    Returns
    -------
    None, updates color_legend_ax
    """
    cmap = plt.get_cmap('viridis')
    import matplotlib.colorbar
    from matplotlib.cm import ScalarMappable

    mappable = ScalarMappable(norm=normalize, cmap=cmap)

    matplotlib.colorbar.Colorbar(
        color_legend_ax, mappable=mappable, orientation='horizontal'
    )

    color_legend_ax.set_title('Log2 mean expression', fontsize=16)

    color_legend_ax.xaxis.set_tick_params(labelsize=20)
    
    
def plot_legend(legend_ax, 
                return_ax_dict, 
                height, 
                min_figure_height, 
                dot_min, 
                dot_max, 
                largest_dot, 
                smallest_dot,
                normalize):

    # to maintain the fixed height size of the legends, a
    # spacer of variable height is added at top and bottom.
    # The structure for the legends is:
    # first row: variable space to keep the first rows of the same size
    # second row: size legend
       
    cbar_legend_height = min_figure_height * 0.08
    size_legend_height = min_figure_height * 0.27
    spacer_height = min_figure_height * 0.3

    height_ratios = [
        height - size_legend_height - cbar_legend_height - spacer_height,
        size_legend_height,
        spacer_height,
        cbar_legend_height,
    ]
    
    fig, legend_gs = make_grid_spec(
        legend_ax, nrows=4, ncols=1, height_ratios=height_ratios
    )

    #size legend
    size_legend_ax = fig.add_subplot(legend_gs[1])
    plot_size_legend(size_legend_ax, dot_min, dot_max, largest_dot, smallest_dot)
    return_ax_dict['size_legend_ax'] = size_legend_ax
    
    #color legend
    color_legend_ax = fig.add_subplot(legend_gs[3])
    plot_colorbar(color_legend_ax, normalize)
    return_ax_dict['color_legend_ax'] = color_legend_ax
    return(return_ax_dict)
    
    
    
```

```python
import matplotlib.gridspec as gridspec
from matplotlib.figure import SubplotParams as sppars, Figure
from typing import Union, List, Sequence, Tuple, Collection, Optional, Callable

def make_grid_spec(
    figsize_or_ax,
    nrows: int,
    ncols: int,
    wspace: Optional[float] = None,
    hspace: Optional[float] = None,
    width_ratios: Optional[Sequence[float]] = None,
    height_ratios: Optional[Sequence[float]] = None,
) -> Tuple[Figure, gridspec.GridSpecBase]:
    kw = dict(
        wspace=wspace,
        hspace=hspace,
        width_ratios=width_ratios,
        height_ratios=height_ratios,
    )
    if isinstance(figsize_or_ax, tuple):
        fig = plt.figure(figsize=figsize_or_ax)
        return fig, gridspec.GridSpec(nrows, ncols, **kw)
    else:
        ax = figsize_or_ax
        ax.axis('off')
        ax.set_frame_on(False)
        ax.set_xticks([])
        ax.set_yticks([])
        return ax.figure, ax.get_subplotspec().subgridspec(nrows, ncols, **kw)


    
def make_figure(df_size, df_color, cell_pair, savepath, save_fig = True):
    category_height = 0.6
    category_width = 0.7
    legends_width = 3
    min_figure_height = 2.5
    
    mainplot_height = len(df_color.index) * category_height
    mainplot_width = len(df_color.columns)* category_width 
    
    height = mainplot_height + 1
    height = max([min_figure_height, height])
    width = mainplot_width + legends_width
    
    return_ax_dict = {}
    legends_width_spacer = 0.7 / width
    
    fig, gs = make_grid_spec(
    (width, height),
    nrows=1,
    ncols=2,
    wspace=legends_width_spacer,
    width_ratios=[mainplot_width , legends_width]
    )
    

    
    
    
    
    # the main plot is divided into three rows and two columns
    # first row is an spacer that is adjusted in case the
    #           legends need more height than the main plot
    # second row is for brackets (if needed),
    # third row is for mainplot and dendrogram/totals (legend goes in gs[0,1]
    # defined earlier)
    
    spacer_height = height - mainplot_height
    height_ratios = [spacer_height, mainplot_height]
    width_ratios = [mainplot_width]
    
    mainplot_gs = gridspec.GridSpecFromSubplotSpec(
        nrows=2,
        ncols=1,
        wspace=0.0,
        hspace=0.0,
        subplot_spec=gs[0, 0],
        width_ratios=width_ratios,
        height_ratios=height_ratios,
    )
    
    #main_plot
    main_ax = fig.add_subplot(mainplot_gs[1, 0])
    dot_max, dot_min, largest_dot, smallest_dot, normalize = dotplot(main_ax, df_size, df_color, cell_pair)
    return_ax_dict['mainplot_ax'] = main_ax
    
    
    #legend_plot
    legend_ax = fig.add_subplot(gs[0, 1])
    print(type(legend_ax))
    return_ax_dict = plot_legend(legend_ax, 
                return_ax_dict, 
                height, 
                min_figure_height, 
                dot_min, 
                dot_max, 
                largest_dot, 
                smallest_dot,
                normalize)
                
    _ax = fig.add_subplot(gs[0, 0])
    _ax.axis('off')
    _ax.set_title(cell_pair, fontsize = 20)
    
    if save_fig == True:
        name = cell_pair.split('|')[0].split('/')[0] + '_' + cell_pair.split('|')[1].split('/')[0] + '.png'
        figpath = os.path.join(savepath, name)
        fig.savefig(figpath, bbox_inches='tight')
        plt.close()

def plot_batch(root_path, group_list, info_matrix, savepath, save_fig):
    #info matrix can be found using function grab_gene_batch in the script cellphone_db_select_pairs
    cell_pair_list = list(info_matrix.index) 
    for cell_pair in cell_pair_list:
        gene_list = literal_eval(info_matrix.loc[cell_pair, 'gene_list']) #change the str back to the list
        dic = table_gen(root_path, group_list, gene_list, cell_pair)
        df_color = dic['log_mean']
        df_size = -np.log10(dic['p_value'] + 0.001)
        
        make_figure(df_size, df_color, cell_pair, savepath,save_fig = save_fig)
    
    
    
    
        
```
