```python
old_names = [x for x in test.columns if '|' in x]
```

```python
df_2 = pd.DataFrame(index = old_names, columns = ['new_names'])
```

```python
for oldname in df_2.index:
    name_1 = oldname.split('|')[0]
    name_2 = oldname.split('|')[1]
    
    newname_1 = df.loc[name_1, 'subtype_new']
    newname_2 = df.loc[name_2, 'subtype_new']
    newname = newname_1+'|'+newname_2
    df_2.loc[oldname, 'new_names'] = newname
    
    
```

```python
df_2
```

```python
#now substitute the names in the output files and store in a new name
root_path = '/domino/edv/id-td-virology/Zhiyuan/public/Gut_remap/processed_data/cellphonedb/'
group_list = ['IT', 'IA', 'HC', 'AR', 'CR']
file_list = ['means.txt', 'significant_means.txt','pvalues.txt']

for group in group_list:
    for file in file_list:
        path = os.path.join(root_path, group, 'out', file)
        temp = pd.read_table(path)
        column_list = [x for x in temp.columns if '|' in x]
        for column_name in column_list:
            newname = df_2.loc[column_name, 'new_names']
            temp.rename(columns={column_name: newname}, inplace = True)
        #write
        filename = file.split('.')[0]+'_new.txt'
        savepath = os.path.join(root_path, group, 'out', filename)
        temp.to_csv(savepath, sep = '\t', index=False)
```
