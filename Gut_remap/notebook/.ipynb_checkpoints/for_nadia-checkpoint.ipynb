```python
def table_gen(root_path, group_list, pair_list, cell_pair):
    dic = {}
    for item in ['out/significant_means.txt', 'out/pvalues.txt']:
        df = pd.DataFrame(0, index = pair_list, columns = group_list)
        for group in group_list:
            path = os.path.join(root_path,group, item) 
            temp = pd.read_table(path)
            temp.set_index('interacting_pair', inplace = True)
            temp = temp.loc[pair_list, cell_pair].to_frame().copy()
            if item == 'out/significant_means.txt':
                temp = temp.fillna(0)
                temp.loc[:,'log_mean'] = np.log1p(temp.loc[:,cell_pair])
                df.loc[pair_list,group] = temp.loc[pair_list, 'log_mean']
            else:
                temp = temp.fillna(1)
                df.loc[pair_list,group] = temp.loc[pair_list, cell_pair]
        if item == 'out/significant_means.txt':
            dic['log_mean'] = df
        else:
            dic['p_value'] = df
    dic['cell_pair'] = cell_pair
    return(dic)
```

```python
group_list = ['IT', 'IA', 'HC', 'CR', 'AR']
root_path = '/domino/edv/id-td-virology/Zhiyuan/public/Gut_remap/processed_data/cellphonedb'
pair_list = ['TNFRSF1A_FASLG',
 'LGALS9_P4HB',
 'LTA_RIPK1',
 'BTLA_TNFRSF14',
 'CLEC2B_KLRF1',
 'HLA-C_FAM3C',
 'CCR6_CCL20',
 'Cholesterol_byDHCR7_RORC',
 'Desmosterol_byDHCR7_NR1H2',
 'LAMP1_FAM3C',
 'Cholesterol_byLIPA_RORC',
 'LGALS9_SLC1A5',
 'LGALS9_SORL1',
 'PDCD1_FAM3C',
 'CXCR3_CCL20',
 'CCL4L2_VSIR',
 'Cholesterol_byLIPA_RORA',
 'FAS_FASLG',
 'LGALS9_CD44',
 'NR3C1_FASLG',
 'Cholesterol_byDHCR7_RORA',
 'TNFSF12_TNFRSF25',
 'TNF_TNFRSF1B',
 'ADORA2A_NAMPT',
 'LTA_TNFRSF1B',
 'TNFSF10_RIPK1',
 'LTA_TNFRSF1A',
 'PGRMC2_CCL4L2']

select= ['LTA_TNFRSF1A',
            'LTA_TNFRSF1B',
            'LTA_RIPK1',
            'LGALS9_CD44',
            'LGALS9_SLC1A5',
            'LGALS9_SORL1',
            'LGALS9_P4HB',
            'PDCD1_FAM3C',
            'LAMP1_FAM3C',
            'HLA-C_FAM3C',
            'FAS_FASLG',
            'NR3C1_FASLG',
            'TNF_TNFRSF1B',
            'TNFSF12_TNFRSF25',
            'TNFSF10_RIPK1',
            'CCR6_CCL20',
            'CXCR3_CCL20',            
            'Cholesterol_byDHCR7_RORC',
            'Cholesterol_byDHCR7_RORA',
            'Cholesterol_byLIPA_RORA',
            'Cholesterol_byLIPA_RORC',
            ]
cell_pair = 'CD4+ Treg CTLA4++|CD8+ MAIT'

dic = table_gen(root_path, group_list, select[::-1], cell_pair)

df_color = dic['log_mean']
df_size = -np.log10(dic['p_value'] + 0.001)
```

```python
savepath = '/domino/edv/id-td-virology/Public_dataset/2022_Gut/for_Nadia/dotplot_color.tsv'
df_color.to_csv(savepath, sep = '\t')
savepath = '/domino/edv/id-td-virology/Public_dataset/2022_Gut/for_Nadia/dotplot_size.tsv'
df_size.to_csv(savepath, sep = '\t')
```
